# Генерация данных для  разной температуры

```python
df = pd.DataFrame()
```

```python

for t in [0+273,10+273,20+273, 25+273]:
    model_i = load_fmu('FinalBat.fmu')
    model_i.set('battery2.T',t)
    res = model_i.simulate( start_time=0.0,final_time=23000.0)
    y_soc=res['battery2.SOC']
    y_volt=res['powerSensor2.voltageSensor.v']
    y_cur=res['powerSensor2.currentSensor.i']
    y_t=res['battery2.T']
    y_t=y_t-273
    y_time=res['time']
    d={'time':res['time'],'voltage':y_volt,'curret': y_cur,'soc':y_soc,'temp':y_t}
    df_1=pd.DataFrame(data=d)
    df=pd.concat([df, df_1])
    df.reset_index(drop=True)
    
    
    
    
    
    
    
```

```python
df.reset_index(drop=True)
```

```python
df.to_csv('out.csv')
```
